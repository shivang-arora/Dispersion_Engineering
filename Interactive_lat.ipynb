```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, IntSlider


def plot_with_slider(i,j):
    x = np.linspace(0, 10, 100)
    
    
    #longitudnal_shift(offset1=grid1[i],  offset2=grid2[j], debug=True)
    ng_slow_bands = 1/np.absolute(np.gradient(bands[i][j], axis=1))/90
    plt.plot(bands[9][9][22], label='Band with no lateral displacement')
    plt.plot(bands[i][j][22])
    slow_band=ng_slow_bands[22]
    
    plt.title(f'Offset 1 = {np.round(grid1[i],3)},    Offset 2 = {np.round(grid2[j],3)}')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.grid(True)
    plt.show()
    plot_band_structure(bands[i][j], band_index=22,zoom=True,freq='frequency')
    plot_group_index(bands[i][j][22],zoom=True,freq='frequency',cut=0)
    
#plt.plot(slow_band)
   
# Create slider widget with integer values
integer_slider = IntSlider(value=9, min=0, max=19, step=1, description='offset1')
integer_slider2 = IntSlider(value=9, min=0, max=19, step=1, description='offset2')
# Create interactive plot with slider
interact(plot_with_slider, i=integer_slider,j=integer_slider2)
```
