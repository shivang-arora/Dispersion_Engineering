```python
import meep as mp
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
import pandas as pd
from func import *
```

```python
def plot_group(freqs,Rs,prominence=0.01,distance=100,a=0.442,N=28):


    peaks, _ = find_peaks(-Rs,prominence=prominence,distance=distance)  # -spectrum to find minima instead of maxima
    plt.plot(1/freqs, Rs, label='Spectrum')
    plt.plot(1/freqs[peaks], Rs[peaks], "x", label='Dips')
    plt.xlim(1.4,1.6)
    dips=freqs[peaks]
    fsr=2*N*a*np.array([dips[i+1]-dips[i] for i in range(len(dips)-1)])/3
    plt.show()
    plt.ylabel('extracted group_index')

    plt.scatter(1/freqs[peaks][:-1],(1/fsr), marker='x')
    plt.xlim(1.4,1.6)
    return freqs[peaks][:-1],1/fsr
```
